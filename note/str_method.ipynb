# ***.strip()***

```python
text=' ：こんちは！　'
print(text.strip())

#半角スペース ( ),全角スペース (　),タブ (\t),改行 (\n, \r)
#人間の目で見た空白は消すよ
```

```python
text='#!#お知らせ!!##'
print(text.strip('#!'))
#引数に入れた文字消すよ
#　端のみ適応
```

# ***.lstrip()  .rstrip()***

```python
text=' colab!'
print(text.lstrip())
print(text.rstrip('!'))
```

# ***.replace()***

```python
text='寿司食べたい'
print(text.replace('寿司','焼肉'))
```

```python
text='2026/05/10/'
print(text.replace('/','-',2))#引数で回数指定
```

# ***.split()***

```python
text='リンゴ,ゴリラ,ラッパ,パナキ'
print(text.split(','))  #文字列をリストに
```

```python
text='リンゴ ゴリラ  ラッパ       パナキ'
print(text.split())  #''空文字はエラー  空白はおｋ
```

# ***.join()***

```python
ls=['ゴリラ','ラッパ','パナキ']
print(','.join(ls))  # '区切りたい文字列'.join(lst) リストから文字列へ
```

```python
ls=['ゴリラ','ラッパ','パナキ']
print('\n'.join(ls))  #改行
print(''.join(ls))  #空文字でぜんぶくっつける
```

# ***f-string***

```python
name='永井博之'
age='45'
print(f'私の名前は{name}で{age}歳です')
```

# ***in***

```python
text='永井博之'
print('博之'in text)
print('浩二'in text)
```

# ***.find()***


```python
st='ゴリラ,ラッパ,パナキ'
print(st.find('パナキ'))  #先頭のインデックスを返す
print(st.find('YAGI'))  #見つからないとー１返す
```

# ***.startswith()  .endwith()***

```python
filename='secret.txt'
print(filename.startswith('sec'))  #冒頭や末尾から文字列サーチ
print(filename.endswith('txt'))
```

# ***.upper()  .lower()***

```python
text='Scikit-learn'
print(text.upper())
print(text.lower())
```

# ***.capitalize()  .title()***

```python
text = "hello-world 123python_programming"
print(text.capitalize()) # 先頭の1文字だけ大文字にする
print(text.title())      # 各単語の先頭を大文字にする
```
