```python
#ループで判定するのは最初の一文字目だけ
#一致したらそこからは t+i のインデックスで連続判定

def simple_match(x, y):
    for t in range(len(x)):
      for i in range(len(y)):
        if x[t+i]!=y[i]:  #インデックス指定、　CATの文字数文ループで同じ文字数送る
          break  #マッチングしなかったらブレイク
      else:
        return t
    return -1

##マッチする文字列があるか判定するのではなく
#マッチしないものを弾く
```
