```python
with open("test.txt", "w") as f:  #ファイル作成
    f.write("こんにちは、製造業")
with open("test.txt", "r") as f:  #ファイル読み込み
    text = f.read()
print(text)
```

```python
with open('seisan.txt','w')as xx:
  xx.write('ライン1: 正常\nライン2: 異常\nライン3: 正常')
with open('seisan.txt','r')as xx:
  y=xx.read()  #一行で出力、\nで改行
print(y)
```

```python
with open("test.txt", "w") as f:
    f.write("こんにちは、製造業")
with open("test.txt", "a") as f:  #追記
    f.write('\n製造業こんにちは')
with open("test.txt", "r") as f:
    text = f.read()
print(text)
```
