# 3-1. 辞書 (dictionary) リファレンス

---

## 📌 辞書とは

**キー (key)** と **値 (value)** を対応づけるデータ構造。

| | 使える型 |
|---|---|
| **キー** | 文字列・数値・タプルなど **変更不可能** なもの（リスト・辞書はNG） |
| **値** | 何でもOK（リスト・辞書も可） |

---
## 🔨 基本操作

```python
# 作成
d = {'apple': 3, 'pen': 5}

# 値の取得
d['apple']          # => 3

# 値の変更・追加（代入でOK）
d['apple'] = 10     # 変更
d['pineapple'] = 7  # 新規追加

# キーの存在確認
'apple' in d        # => True
'banana' in d       # => False

# 要素数
len(d)              # => キーと値のペアの数

# 削除
del d['pen']        # キーを指定して削除

# 空の辞書
empty = {}
```

> ⚠️ **注意点**
> - 存在しないキーを `d[key]` で取得すると **KeyError**
> - 値からキーは直接取得できない
> - 数字インデックス `d[0]` は使えない（キーが整数の場合を除く）

---
## 🔧 主要メソッド

```python
d = {'apple': 3, 'pen': 5}

# get(key)         存在しないキーでもエラーにならない → None を返す
d.get('apple')          # => 3
d.get('orange')         # => None
d.get('orange', -1)     # => -1（デフォルト値を指定できる）

# keys() / values() / items()   いずれも「リストのようなもの」を返す
list(d.keys())          # => ['apple', 'pen']
list(d.values())        # => [3, 5]
list(d.items())         # => [('apple', 3), ('pen', 5)]

# pop(key)         削除 + 削除した値を返す
d.pop('pen')            # => 5  (辞書からも削除される)

# clear()          全要素を削除 → 空の辞書になる
d.clear()

# copy()           辞書を複製（片方を変更しても影響しない）
d2 = d.copy()
```

### ▲ setdefault — キーがない場合だけ登録

```python
d = {'apple': 3, 'pen': 5}

# キーがある → 既存の値を返すだけ（辞書は変わらない）
d.setdefault('apple', 99)   # => 3

# キーがない → 値を登録して返す
d.setdefault('orange', 99)  # => 99  (d に {'orange': 99} が追加される)

# ↑ 上記は以下と同じ意味
if 'orange' not in d:
    d['orange'] = 99
```

### ▲ keys/values/items の返値は「動的」に変わる

```python
d = {'apple': 3, 'pen': 5}
ks = d.keys()       # この時点では ['apple', 'pen']
d['kiwi'] = 9
list(ks)            # => ['apple', 'pen', 'kiwi']  ← 辞書の変更が反映される！
# ※ list() に変換した瞬間にスナップショットになる
```

---
## 🗂️ 辞書とリストの組み合わせ

```python
# 値にリストを持つ辞書
numbers = {'dozens': [10, 20, 40], 'hundreds': [100, 101, 120]}
numbers['dozens']       # => [10, 20, 40]
numbers['dozens'][1]    # => 20

# 辞書を要素に持つリスト
ppap = {'apple': 3, 'pen': 5}
pets = {'cat': 3, 'dog': 3}
ld = [ppap, pets]
ld[1]           # => {'cat': 3, 'dog': 3}
ld[1]['dog']    # => 3
```

---
## 📋 メソッド一覧

| メソッド | 動作 | 戻り値 |
|---|---|---|
| `d[key]` | 値の取得 | 値（なければ KeyError） |
| `d.get(key, default)` | 値の取得（安全） | 値 or None or default |
| `d.keys()` | キー一覧 | dict_keys（動的） |
| `d.values()` | 値一覧 | dict_values（動的） |
| `d.items()` | キー・値のタプル一覧 | dict_items（動的） |
| `d.pop(key)` | 削除して値を返す | 削除した値 |
| `d.clear()` | 全削除 | None |
| `d.copy()` | 複製 | 新しい辞書 |
| `d.setdefault(key, val)` | なければ登録 | 既存 or val |
| `del d[key]` | キー削除 | — |
| `key in d` | キーの存在確認 | True / False |
| `len(d)` | 要素数 | int |
