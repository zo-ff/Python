### 練習 5：総合演習（make_index_dict）
リスト `ls` が与えられたとき、**「値」をキー、「インデックス」を値**とした辞書を返す関数 `make_index_dict` を作成してください。

```python
def make_index_dict(ls):
    index_dict = {}
    for i, value in enumerate(ls):
        index_dict[value] = i
    return index_dict

# テストコード
data = ["apple", "pen", "orange"]
result = make_index_dict(data)
print(result == {'apple': 0, 'pen': 1, 'orange': 2})
print(f"作成された辞書: {result}")
```
