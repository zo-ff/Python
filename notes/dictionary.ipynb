```python

d={'A':100,'B':10,'C':9999}
```

---
***d.keys()***

keyを返す

```python
print(d.keys())
```

---
***d.values()***

valueを返す

```python
print(d.values())
```

---
***d.items()***

keyとvalueを返す


```python
print(d.items())
```

---
***d.get()***

指定したkeyのvalueを返す

存在しないkeyでもエラーにならず実務向き、デフォ値指定可

```python
print(d.get('A'))
print(d.get('D'))
print(d.get('D',0))
```

---
💣***d.update({})***

上書き更新、dict形式で渡す

```python
d.update({'D':1})
print(d)
```

---
💣***d.pop()***

削除、keyのみでおｋ

```python
d.pop('D')
print(d)
```
